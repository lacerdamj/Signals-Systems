```python
import numpy as np
import matplotlib.pyplot as plt
```

### Time-scaling transformation

Consider the following signal

$$x(t)= \left\{ \begin{aligned} 
3t & \quad 0 \leq t \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. $$

```python
n = 10000
tf = 6
t = np.linspace(0, tf, n)

x = np.where(np.logical_and(t>=0, t<=2), 3*t, 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,x,'b', label="$x$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x(t)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

The signal $x(2t)$ reads

$$x(2t)= \left\{ \begin{aligned} 
3\times 2t & \quad 0 \leq 2t \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. = \left\{ \begin{aligned} 
6t & \quad 0 \leq t \leq 1 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. $$

This signal $x(2t)$ will compress the original signal $x(t)$ with a factor $2$, as shown in the next Figure.

```python

x1 = np.where(np.logical_and(2*t>=0, 2*t<=2), 3*(2*t), 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,x1, 'r', label="$x(2t)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x(2t)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

The signal $x(\frac{t}{2})$ reads

$$x(\frac{t}{2})= \left\{ \begin{aligned} 
3\times \frac{t}{2} & \quad 0 \leq \frac{t}{2} \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. = \left\{ \begin{aligned} 
1.5t & \quad 0 \leq t \leq 4 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. $$

The signal $x(\frac{t}{2})$ will expand the original signal $x(t)$ with a factor $2$, as shown in the next Figure.

```python
x2 = np.where(np.logical_and((t/2)>=0, (t/2)<=2), 3*(t/2), 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,x2, 'k', label="$x(t/2)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x(t/2)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

```python

fg, ax = plt.subplots(1, 1)

ax.plot(t,x, 'b', label="$x(t)$", linewidth=2)
ax.plot(t,x1, 'r', label="$x(2t)$", linewidth=2)
ax.plot(t,x2, 'k', label="$x(t/2)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

### Time-shift transformation. 

Consider the following signal

$$f(t)= \left\{ \begin{aligned} 
t^2-1 & \quad 1 \leq t \leq 3 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. $$ 

```python
f = np.where(np.logical_and(t>=1, t<=3), t**2-1, 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,f,'b', label="$f(t)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$f(t)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

To plot the signal $f(t-1)$, you must change $t$ by $t-1$ everywhere in the original function. In this case we have

$$f(t-1)= \left\{ \begin{aligned} 
(t-1)^2-1 & \quad 1 \leq t-1 \leq 3 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. = \left\{ \begin{aligned} 
t^2-2t & \quad 2 \leq t \leq 4 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right.$$

Note that shifting the function, the intervals in which it is evaluated also change. As a result the original function will shift to the right in this case, as shown in the next Figure.

```python
f1 = np.where(np.logical_and(t-1>=1, t-1<=3), (t-1)**2-1, 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,f1,'orange', label="$f(t-1)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$f(t-1)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

To plot the signal $f(t+1)$, you must change $t$ by $t+1$ in the original function. In this case we have

$$f(t+1)= \left\{ \begin{aligned} 
(t+1)^2-1 & \quad 1 \leq t+1 \leq 3 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. = \left\{ \begin{aligned} 
t^2+2t & \quad 0 \leq t \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right.$$ 

In this scenario the original function will shift to the left, as shown in the next Figure.

```python
f2 = np.where(np.logical_and(t+1>=1, t+1<=3), (t+1)**2-1, 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,f2,'brown', label="$f(t+1)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$f(t+1)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

In the next plot we can see $f(t)$, $f(t-1)$, $f(t+1)$. 

```python
fg, ax = plt.subplots(1, 1)

ax.plot(t,f, 'blue', label="$f(t)$", linewidth=2)
ax.plot(t,f1, 'orange', label="$f(t-1)$", linewidth=2)
ax.plot(t,f2, 'brown', label="$f(t+1)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

### Time-reversal transformation. 

Consider the following signal

$$g(t)= \left\{ \begin{aligned} 
e^{-2t} & \quad 0 \leq t \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. $$ 

The signal $g(-t)$ is the mirrored image of the function $g(t)$ with respect to the vertical axis. The signal $g(-t)$ can be obtained as follows 

$$g(-t)= \left\{ \begin{aligned} 
e^{-2(-t)} & \quad 0 \leq -t \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. = 
\left\{ \begin{aligned} 
e^{2t} & \quad -2 \leq t \leq 0 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right.$$ 

In the sequel the figures show both signals. 


```python
n = 10000
tf = 3
t = np.linspace(-3, tf, n)

g = np.where(np.logical_and(t>=0, t<=2), np.exp(-2*t), 0) 

gm = np.where(np.logical_and(-t>=0, -t<=2), np.exp(-2*(-t)), 0) 

# plot results:    

fg, (ax1, ax2) = plt.subplots(2)

fg.set_size_inches(9, 9)

ax1.plot(t,g,'b', label="$g(t)$", linewidth=2)
ax1.grid(True)
ax1.legend(loc='best',fontsize = 12.0)
ax1.set_ylabel('g(t)', fontsize = 12.0)
ax1.set_xlabel('t', fontsize = 12.0)


ax2.plot(t,gm,'k--', label="$g(-t)$", linewidth=2)
ax2.grid(True)
ax2.legend(loc='best',fontsize = 12.0)
ax2.set_ylabel('g(-t)', fontsize = 12.0)
ax2.set_xlabel('t', fontsize = 12.0)

fg.canvas.draw()
```

### Combined transformation

For instance, consider the signal

$$
f(t)= \left\{ \begin{aligned} 
t^2-1 & \quad 1 \leq t \leq 3 \\
0 & \quad \mbox{otherwise.} 
\end{aligned} \right. 
$$ 

Obtain the signal $f(2t+1)$. In this case, you must shift the original function to the left, and compress the resulting signal by a factor of two. 


$$
f(2t+1)= \left\{ \begin{aligned} 
(2t+1)^2-1 & \quad 1 \leq 2t+1 \leq 3 \\
0 & \quad \mbox{otherwise.} 
\end{aligned} \right. =
\left\{ \begin{aligned} 
(2t+1)^2-1 & \quad 0 \leq 2t \leq 2 \\
0 & \quad \mbox{otherwise.} 
\end{aligned} \right. =
\left\{ \begin{aligned} 
(2t+1)^2-1 & \quad 0 \leq t \leq 1 \\
0 & \quad \mbox{otherwise.} 
\end{aligned} \right.
$$ 

The next figure shows the original signal $f(t)$, and the signal $f(2t+1)$.

```python
n = 10000
tf = 6
t = np.linspace(0, tf, n)

r = np.where(np.logical_and(2*t+1>=1, 2*t+1<=3), (2*t+1)**2-1, 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,r,'green', label="$f(2t+1)$", linewidth=2)
ax.plot(t,f,'blue', label="$f(t)$", linewidth=2)

ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```
