```python
#Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

### Mudança de escala de tempo

Considere o sinal 

$$x(t)= \left\{ \begin{aligned} 
3t & \quad 0 \leq t \leq 2 \\
0 & \quad \mbox{otherwise} 
\end{aligned} \right. $$

```python
n = 10000   #número de pontos
t0 = 0        #tempo inicial
tf = 6      #tempo final
#a função linspace define um intervalo de t0 até tf 
#igualmente espaçado em n pontos
t = np.linspace(t0, tf, n)

# define a função x(t)
x = np.where(np.logical_and(t>=0, t<=2), 3*t, 0) 

# inicia a figura
fg, ax = plt.subplots(1, 1)
#plota no eixo horizontal a variável t e no vertical a variável x
# 'b' de blue define a cor azul.
ax.plot(t,x,'b', label="$x$", linewidth=2)
#insere uma grade no gráfico
ax.grid(True)
#posiciona a legenda na melhor localização
ax.legend(loc='best')
#insere informação do eixo vertical
plt.ylabel('$x(t)$', fontsize=12)
#insere informação do eixo horizontal
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

O sinal $x(2t)$ pode ser obtido como

$$x(2t)= \left\{ \begin{aligned} 
3\times 2t & \quad 0 \leq 2t \leq 2 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. = \left\{ \begin{aligned} 
6t & \quad 0 \leq t \leq 1 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. $$

O sinal $x(2t)$ vai comprimir o sinal original $x(t)$ por um fator $2$, como apresentado na próxima Figura.

```python
# define a função x(2t)
x1 = np.where(np.logical_and(2*t>=0, 2*t<=2), 3*(2*t), 0) 

fg, ax = plt.subplots(1, 1)
ax.plot(t,x1, 'r', label="$x(2t)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x(2t)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Para o sinal $x(\frac{t}{2})$ tem-se

$$x(\frac{t}{2})= \left\{ \begin{aligned} 
3\times \frac{t}{2} & \quad 0 \leq \frac{t}{2} \leq 2 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. = \left\{ \begin{aligned} 
1.5t & \quad 0 \leq t \leq 4 \\
0 & \quad \mbox{caso contrário} 
\end{aligned} \right. $$

O sinal $x(\frac{t}{2})$ vai expandir o sinal original $x(t)$ com um fator $2$, como ilustrado na próxima Figura.

```python
# define a função x(t/2)
x2 = np.where(np.logical_and((t/2)>=0, (t/2)<=2), 3*(t/2), 0) 

fg, ax = plt.subplots(1, 1)
ax.plot(t,x2, 'k', label="$x(t/2)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x(t/2)$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```

Por fim, vamos plotar todos os gráficos em uma única figura para visualizar o efeito da mudança de escala de tempo no sinal $x(t)$.

```python
fg, ax = plt.subplots(1, 1)

ax.plot(t,x, 'b', label="$x(t)$", linewidth=2)
ax.plot(t,x1, 'r', label="$x(2t)$", linewidth=2)
ax.plot(t,x2, 'k', label="$x(t/2)$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.xlabel('t', fontsize=12)
fg.canvas.draw()
```
