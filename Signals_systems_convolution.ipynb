```python


import numpy as np
import matplotlib.pyplot as plt
```

##Example 1
Compute the convolution $x_1(t)\ast x_2(t)$ considering
$$x_1(t)=t(u(t)-u(t-3)),$$
and
$$x_2(t)=4(u(t-1)-u(t-2)).$$




```python
n = 10000
tf = 6
t = np.linspace(0, tf, n)
T = t[1] - t[0]  # sampling width

x1 = np.where(np.logical_and(t>=0, t<=3), t, 0) # build input functions

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,x1, label="$x_1$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x_1$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()

```

```python
x2 = np.where(np.logical_and(t>=1, t<=2), 4, 0)
fg, ax = plt.subplots(1, 1)

ax.plot(t,x2,'r', label="$x_2$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$x_2$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()

```

```python
y = np.convolve(x1, x2, mode='full')*T   # scaled convolution 
# the function is substituting an integration for a summation, for this reason we must include the value T,
# which makes the role of dtau in the integral.

ty = np.linspace(0, 2*tf, n*2-1)  # double time interval

# plot results:    
fg, ax = plt.subplots(1, 1)
ax.plot(ty, y, 'g', label="$x_1\\ast x_2$", linewidth=2)
ax.legend(loc='best')
plt.ylabel('$x_1\\star x_2$', fontsize=12)
plt.xlabel('t', fontsize=12)
plt.title('Convolution', fontsize=12)
ax.grid(True)

fg.canvas.draw()
```


Note that, the width of the convolved signal $x_1\ast x_2$ is $4$, which is the sum of the width of the signal $x_1$, with the width of the signal $x_2$.

##Example 2
Compute the convolution $g(t)\ast h(t)$ considering
$$g(t)=t(u(t)-u(t-1))+(2-t)(u(t-1)-u(t-2)),$$
and
$$h(t)=u(t)-2u(t-1)+u(t-2).$$

```python
g1 = np.where(np.logical_and(t>=0, t<=1), t, 0)  # build input functions
g2 = np.where(np.logical_and(t>=1, t<=2), 2-t, 0)  # build input functions
g = g1+g2 

# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,g, 'k', label="$g$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$g$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()

```

```python
h1 = np.where(np.logical_and(t>=0, t<=1), 1, 0)
h2 = np.where(np.logical_and(t>=1, t<=2), -1, 0)
h = h1+h2


# plot results:    
fg, ax = plt.subplots(1, 1)

ax.plot(t,h, 'm', label="$h$", linewidth=2)
ax.grid(True)
ax.legend(loc='best')
plt.ylabel('$h$', fontsize=12)
plt.xlabel('t', fontsize=12)
fg.canvas.draw()


```

```python
y = np.convolve(g, h, mode='full')*T   # scaled convolution 
ty = np.linspace(0, 2*tf, n*2-1)  # double time interval

# plot results:    
fg, ax = plt.subplots(1, 1)
ax.plot(ty, y, 'c', label="$g \\ast h$")
ax.legend(loc='best')
plt.ylabel('$g\\ast h$', fontsize=12)
plt.xlabel('t', fontsize=12)
plt.title('Convolution', fontsize=12)
ax.grid(True)

fg.canvas.draw()
```

The convolution resulted in a signal with width $4$, which is the sum of the width of the signal $g$ with the width of the signal $h$.
